# Buy & Hold Backtest Report

Run a backtest using the Buy & Hold strategy, against a stock using the Alpaca 
Market Data API and use an LLM to get a summary report of the run.

### Setup Environment

```python
# Load environment variables from .env file
%reload_ext dotenv
%dotenv ../.env
```

### Get Recent Bars

Using the Alpaca market data API get recent (60 days) candlestick data for AAPL
with the timeframe of 15 minutes.

```python
from datetime import timedelta
from qalc.platform.market_data import TimeFrame, AssetClass
from qalc.platform.market_data.providers import AlpacaDataApi

# Get the bars for the last 60 days
data_provider = AlpacaDataApi(symbol="AAPL", asset_class=AssetClass.STOCK)
bars = data_provider.get_bars_window(
    window=timedelta(days=60), 
    timeframe=TimeFrame.FIFTEEN_MINUTES
)
```

```python
# Display the bars
display(bars)
```

### Execute Strategy

Using the recent AAPL candlestick data backtest the simple buy & hold strategy 
by way of example.

```python
from qalc.strategies.buy_and_hold import BuyAndHoldStrategy

# Set up Buy and Hold strategy
buy_and_hold = BuyAndHoldStrategy()
results = buy_and_hold.backtest(bars, cash=100000)
```

### Generate Summary Report

Using an LLM, generate a summary report of the backtest results and render in
Markdown format.

```python
from qalc.llm import llm_init
from qalc.llm.prompts import get_prompt

llm = llm_init(model="gpt-3.5-turbo", temperature=0.5)
prompt = get_prompt("backtest_summary", 
    strategy_name="buy_and_hold",
    results=results
)
report = llm.invoke(prompt)
```

```python
# Display the generated report
from IPython.display import Markdown
display(Markdown(report.content))
```
